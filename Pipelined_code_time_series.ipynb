```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor

```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/fazalpge/AI-based-Smart-ERP-Next/main/Inventory%20Data.csv")  # Replace "your_dataset.csv" with the path to your dataset
```

```python
df.head()
```

```python
data=df.drop('TransactionNo', axis=1)
data
```

```python
df.describe()
```

```python
# Convert 'TransactionDate' column to datetime
df ["TransactionDate"]= pd.to_datetime(df['TransactionDate'])
```

```python
df.isnull().sum()
```

```python
df=df.fillna(method='pad',axis=1)
```

```python
df.info()
```

```python
# Sort dataframe by 'TransactionDate' for time series data
#df = df.sort_values(by='TransactionDate')

```

```python
# Define features and target variable
X = df[['TransactionDate', 'Cost', 'SaleBonusQty', 'OtherCharge', 'VATAmount']]
y = df['SaleQty']
```

```python
X
```

```python
from sklearn.preprocessing import FunctionTransformer

import numpy as np

# Function to extract datetime features
def extract_datetime_features(X):
    return np.vstack([[x.year, x.month, x.day, x.hour, x.minute, x.second] for x in X])



# Define the transformer
date_transformer = FunctionTransformer(extract_datetime_features, validate=False)

# Define the preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), slice(1,5)),  # numeric_features contains indices/column names of numerical features
        ('date', date_transformer, 0)   # datetime_features contains indices/column names of datetime features
    ],
    remainder='drop'  # Drop any columns not explicitly transformed
)
```

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
X_train.info()
```

```python
X_train
```

```python
preprocessor.fit_transform(X_train)
```

```python
preprocessor_df=pd.DataFrame(preprocessor.fit_transform(X_train))
```

```python
preprocessor_df
```

```python
model = XGBRegressor()
```

```python
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', model)
])
```

```python
# Fit the pipeline
pipeline.fit(X_train, y_train)
```

```python
# Make predictions
predictions = pipeline.predict(X_test)
```

```python
# Create a new DataFrame for predictions with weekly or monthly basis
predictions_df = pd.DataFrame({
    'Date': X_test['TransactionDate'],  # Assuming 'TransactionDate' is the column containing date
    'Prediction': predictions
})
```

```python
predictions_df
```

```python
# Aggregate predictions on a weekly basis
weekly_predictions = predictions_df.resample('W-Mon', on='Date').sum()
weekly_predictions
```

```python
# Aggregate predictions on a monthly basis
monthly_predictions = predictions_df.resample('M', on='Date').sum()
monthly_predictions
```

```python
# Aggregate predictions on a half_yearly basis
half_yearly_predictions = predictions_df.resample('6M', on='Date').sum()
half_yearly_predictions
```

```python
# Aggregate predictions on a yearly basis
yearly_predictions = predictions_df.resample('Y', on='Date').sum()
yearly_predictions
```

```python
# Print the predictions
print("Weekly Predictions:")
print(weekly_predictions)

print("\nMonthly Predictions:")
print(monthly_predictions)

print("\nhalf_yearly_predictions:")
print(half_yearly_predictions)

print("\nyearly_predictions:")
print(yearly_predictions)
```
